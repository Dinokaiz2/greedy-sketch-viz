```python
#Points sampled from 5 toruses of varying sizes with a larger spread in 'c' and low values in 'a'
c = [11,25,100, 250, 500]
a = range(5)

points1 = np.empty((0,3))
for i in range(5):
    points1 = np.append(points1, tadasets.torus(n=400, c=c[i], a=a[i]), axis=0)

print(points1.shape)
fig1 = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.set_zlim(bottom=-400, top=400)
ax1.scatter(points1.T[0], points1.T[1], points1.T[2])
```

```python
rips1= ripser.ripser(points1)
persim.plot_diagrams(rips1['dgms'], show=True)
```

```python
#Similar to the previous cell but 'a' values are a lot higher now and 'c' values have a much lower spread
c2 = [11,25,50,75,100,200,250,300,400,500]
a2 = [5,10,25,50,75,100,125,150,175,200]

points2 = np.empty((0,3))
for i in range(10):
    points2 = np.append(points2, tadasets.torus(n=200, c=c2[i], a=a2[i]), axis=0)

print(points2.shape)
fig2 = plt.figure()
ax2= plt.axes(projection='3d')
ax2.set_zlim(bottom=-400, top=400)
ax2.scatter(points2.T[0], points2.T[1], points2.T[2])
```

```python
rips2= ripser.ripser(points2)
persim.plot_diagrams(rips2['dgms'], show=True, xy_range=[0,300,0,300])
```
